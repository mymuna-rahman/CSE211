#**Stack & Queue**

[Python Stacks, Queues, and Priority Queues in Practice](https://realpython.com/queue-in-python/)

**Stack**

```python
sample_stack = ['alpha', 'beta', 'gamma', 'delta']

print(sample_stack)
```

```python
#push into stack

sample_stack.append('epsilon')

sample_stack.append('zeta')

sample_stack.append('eta')

print(sample_stack)
```

```python
#pop from stack

sample_stack.pop()

sample_stack.pop()

print(sample_stack)
```

**Queue**

```python
sample_queue = ['alpha', 'beta', 'gamma', 'delta']

print(sample_queue)
```

```python
#push into queue

sample_queue.append('epsilon')

sample_queue.append('zeta')

sample_queue.append('eta')

print(sample_queue)
```

```python
#deque from queue

sample_queue.pop(0)

sample_queue.pop(0)

sample_queue.pop(0)

print(sample_queue)
```
